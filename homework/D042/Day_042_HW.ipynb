```python
from sklearn import datasets, metrics
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split

import pandas as np
```

```python
boston = datasets.load_boston()

df = np.DataFrame(boston.data)
df.columns = boston.feature_names
df2 = np.DataFrame(boston.target)
df = np.concat([df, df2], axis = 1)
df = df.rename(columns={0:"target"})
df.head(20)
```

```python
# 切資料集
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size = 0.3, random_state = 3)
# 建立模型
csf = DecisionTreeRegressor()
# TRAIN
csf.fit(x_train, y_train)
# PREDICT
y_pred = csf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
acc = metrics.r2_score(y_test, y_pred)
print("Rsquare:",acc)
mse = metrics.mean_squared_error(y_test, y_pred)
print("MSE:",mse)
```
